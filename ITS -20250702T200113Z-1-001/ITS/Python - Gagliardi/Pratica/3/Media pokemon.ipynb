Gestire i pokemon attraverso la programmzione ad oggetti. <br>
Visualizzare le caratteristiche dei pokemon e confrontarli attraverso le medie

```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file CSV dei Pokémon
url = '/content/drive/MyDrive/Python - Gagliardi/Pratica/3/Pokemon.csv'
df = pd.read_csv(url , sep=',' , encoding = 'latin1')
#display(df)

class Pokecenter():
  def pokemon1(self, df):
    id = int(input("Inserisci il numero del pokemon: "))
    pokeinfo = df.loc[id, ["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]]
    print("Pokemon selezionato: ", df.loc[id, "name"])
    somma = pokeinfo.sum()
    med1 = np.mean(pokeinfo)
    print("Media stats:", med1)
    return med1

  def pokemon2(self, df):
    id = int(input("Inserisci il numero del pokemon: "))
    pokeinfo = df.loc[id, ["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]]
    print("Pokemon selezionato: ", df.loc[id, "name"])
    somma = pokeinfo.sum()
    med2 = np.mean(pokeinfo)
    print("Media stats:", med2)
    return med2


class Studio:
    def graph(self, med1, med2):
        labels = ['Pokemon 1', 'Pokemon 2']
        medie = [med1, med2]

        plt.bar(labels, medie, width=0.3, color=['blue', 'red'])
        plt.xlabel('Giocatori')
        plt.ylabel('Media Statistiche')
        plt.title('Confronto delle Medie delle Statistiche dei Pokémon')
        plt.show()

#chiamo le classi in variabili
pokecenter = Pokecenter()
studio = Studio()

#chiamo le medie dalle due funzioni
med1 = pokecenter.pokemon1(df)
med2 = pokecenter.pokemon2(df)

studio.graph(med1, med2)

if med1 > med2:
  print("Pokemon 1 vince la battaglia!")
elif med1 < med2:
  print("Pokemon 2 vince la battaglia!")
else:
  print("È un pareggio!")
```
