# **ACCESSO AD UN DATAFRAME**

**By position**

**`iloc[[0], [2]]`**

 restituisce la riga e la colonna alla posizione indicata

**`[0]`** il primo indice accede alle righe

**`[2]`** il secondo indice accede alle colonne

```python
tmp_val = df_auto.iloc[[0], [2]]
display(tmp_val)
```

**At position**

**`iat[0, 2]`**

 restituisce solo il singolo valore corrispondente a riga e colonna indicate

  **`0`** il primo indice accede agli indici

 **`2`** il secondo indice accede alle colonne

```python
display(df_auto.iat[0, 2])
```

**By label**

**`loc[[2], ["Circonferenza"]]`**

Restituisce  la riga 2 nella colonna "Circonferenza"

```python
display(df_trees.loc[[2], ["Circonferenza"]])
```

**At label**

**`loc[[2], ["Circonferenza"]]`**

Restituisce il valore individuato dalla riga 2 nella colonna "Circonferenza"

```python
display(df_trees.at[2, "Circonferenza"])
```

**By column**

restituisce l'intero dontenuto di una o pi√π colonne

```python
display(df_trees[["Id", "Circonferenza"]])
```

**By row**

Restituisce tutti i termini di una riga

```python
display(df_trees.iloc[0])
```

**Bolean indexing**

Filtra un dataset attraverso delle maschere

```python
# crea una maschera di selezione che esclude tutte le righe che hanno i valori  minori di 20,
# in corrispondenza della colonna "Circonferenza"
mask_1 = df_trees["Circonferenza"] < 12

display(mask_1)
```

```python
# filtra il dataset utilizzando una maschera
df_mask_1 = df_trees[mask_1]

display(df_mask_1)
```

# **MODIFICA DEI DATAFRAME**

```python
df1 = pd.DataFrame({"Col1": [1, 2, 3],
                    "Col2": ["A", "B", "C"],
                    "Col3": ["Ford", "Fiat", "Opel"]})


df2 = pd.DataFrame({"C1": [4, 5, 6],
                    "C2": ["D", "E", "F"]})
```

**Joint di DataFrame**

```python
# aggiungere una colonVolumena di un DataFrame ad un altro
df_j = df2.join(df1['Col3'])

display(df_j)
```

**Concatenazione di DataFrame**

```python
df_c = pd.concat([df1, df2], axis=1)
display(df_c)
```

**Ordinamento**

```python

data = {'Name': ['Ben', 'Anna', 'Zoe', 'Tom', 'John', 'Steve', 'Becky', 'Bob'],
            'Score':  [1, 4, 3, 2, 1, 1, 2, 2],
            'Age': [20, 27, 20, 43, 30, 20, 22, 21]}

df_names = pd.DataFrame(data)

display(df_names)
```

**`sort_values()`**

Ordina gli elementi di un DataFrame usando l'etichetta di una colonna.


**`by `**

indica la colonna da usare come maschera



**`inplace `**

indica se il dataframe di partenza deve essere definitivamente modificato

```python
df_names.sort_values( by=['Age'],  inplace=True)

display(df_names)
```

**`sort_index()`**

Ordina gli elementi di un DataFrame attraverso gli indici delle righe.

```python
copy_fn_names = df_names.sort_index()
display(copy_fn_names)
```
